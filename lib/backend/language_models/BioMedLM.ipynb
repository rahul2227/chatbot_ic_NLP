```python
# Load model directly
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
tokenizer = GPT2Tokenizer.from_pretrained("stanford-crfm/BioMedLM")

model = GPT2LMHeadModel.from_pretrained("stanford-crfm/BioMedLM")
```

```python
device = "cpu"

# Set padding token to eos_token
tokenizer.pad_token = tokenizer.eos_token

# Encode input text with attention mask
input_text = "Breast Cancer is"
input_ids = tokenizer.encode(input_text, return_tensors="pt", padding=True, truncation=True).to(device)


# Generate text in batches
batch_size = 4
generated_texts = []
for i in range(0, len(input_ids), batch_size):
    batch_output = model.generate(input_ids[i:i+batch_size], do_sample=True, max_length=100)
    generated_texts.extend([tokenizer.decode(output, skip_special_tokens=True) for output in batch_output])

# Decode and print generated output
print("Output:\n" + 100 * "-")
for i, text in enumerate(generated_texts):
    print(f"Generated text {i + 1}:\n{text}\n")
import torch

device = torch.device("cuda")

# Encode input text with attention mask
input_text = "Breast Cancer is"
input_ids = tokenizer.encode(input_text, return_tensors="pt", padding=True, truncation=True).to(device)

# Set pad_token_id to eos_token_id for open-end generation
model.config.pad_token_id = model.config.eos_token_id

# Generate text
sample_output = model.generate(input_ids, do_sample=True, max_length=50, top_k=50)

# Decode and print generated output
print("Output:\n" + 100 * "-")
print(tokenizer.decode(sample_output[0], skip_special_tokens=True))
```
